```python
def mini_quiz_app():
    sorular = [
        {"soru":"(5+3)/2=?", "cevap":"4"},
        {"soru":"Türkiye'nin başkenti neresidir?", "cevap":"ankara"},
        {"soru":"Türkiye'nin kaç ili vardır?", "cevap":"81"},
        {"soru":"Türkiye'nin en büyük gölü hangisidir?", "cevap":"van gölü"},
        {"soru":"Türkiye'nin en yüksek dağı hangisidir?", "cevap":"ağrı dağı"},
        {"soru":"Türkiye'nin en uzun nehri hangisidir?", "cevap":"kızılırmak"},
        {"soru":"Türkiye'nin en kalabalık şehri hangisidir?", "cevap":"istanbul"}
    ] # sorularımızı liste içerisinde her biri sözlük olacak şekilde tanımladık.
    puan = 0 # kullanıcının başlangıç puanı 0 doğru cevap verdikçe 1 puan alacak.
    for sira, soru in enumerate(sorular, start=1): # kullanıcıya soruları sırayla göstermek için sorularımızı sırayla gezeceğiz.
        print(f"{sira}.Soru: {soru['soru']}") # for döngüsü ile ilk soruyu kullanıcıya interaktif olarak soruyoruz.
        user_answer = input("Yukarıdaki sorunun cevabını yazınız: ").strip().lower() # kullanıcının cevabını alıyoruz baştaki ve sondaki boşlukları temizleyip küçük harfe çeviriyoruz.
        if user_answer == soru["cevap"]: # kullanıcının cevabını kontrol ediyoruz. Eğer cevap doğruysa kullanıcıyı tebrik edip puanını 1 artırıyoruz.
            print("Tebrikler doğru cevap!")
            puan += 1
        else: # kullanıcı soruya yanlış cevap verdiyse kullanıcıya sorunun cevabını öğrenmesi için doğru cevabı gösteriyoruz.
            print(f"Maalesef yanlış cevap verdiniz. Doğru cevap: {soru['cevap']}")
        # her soru bitine kadar soruları for döngüsü ile sırayla gezdik.
    print(f"Quiz tamamlandı! Toplam almış olduğunuz puan: {puan}/{len(sorular)}") # Şimdi kullanıcıya toplam soru sayısı üzerinden aldığı puanı gösteriyoruz. Doğru sayısı / Toplam Soru sayısı
```

```python
mini_quiz_app()
```
