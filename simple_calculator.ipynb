### Proje 2-Simple Calculator

Yapılacaklar
* Kullanıcıdan iki sayı al
* Kullanıcıdan işlem seçmesini iste (toplama, çıkarma, çarpma, bölme)
* Sonucu ekrana yazdır
* Sıfıra bölme hatasını yakala

```python
def calculator():    
    try:
        first_number = float(input("İlk sayıyı girin: ")) # birinci sayi
        second_number = float(input("İkinci sayıyı girin: ")) # ikinci sayi
        operator = input("İşlem seçin (+, -, *, /): ") # islem secimi 
        print()
        if operator == "+":
            print(f"Sonuç: {first_number + second_number}")
        elif operator == "-":
            print(f"Sonuç: {first_number - second_number}")
        elif operator == "*":
            print(f"Sonuç: {first_number * second_number}")
        elif operator == "/":
            print(f"Sonuç: {first_number / second_number}")
        else:
            print("Hata: Geçersiz işlem operatörü.")
    except ZeroDivisionError:
        print("Hata: Bölme işleminde bölünen 0 olamaz.")
    except ValueError:
        print("Hata: Lütfen geçerli bir sayı girin.")

calculator()
```
