### Proje 1-Hello World Customizer

Yapılacaklar
* Kullanıcıdan isim al
* Kaç kez yazdırmak istediğini al
* Mesajı o kadar kez ekrana yazdır

```python
def greet_user():
    user_name = input("Lütfen isminizi giriniz: ") # mesaj kişiye özel olacak. 
    frequency = int(input("İsminizi kaç kez yazdırmak istiyorsunuz?")) # input (string) tipini int e çeviriyoruz for döngüsünde hata almamak için.
    print() # programın güzel gözükmesi için 1 boşluk
    for i in range(frequency): # kod tekrarını önlemek için aynı işlemde for. DRY prensibi
        print(f"Merhaba, {user_name}!")

greet_user()
```
