### Proje 1-Hello World Customizer

Yapılacaklar
* Kullanıcıdan isim al
* Kaç kez yazdırmak istediğini al
* Mesajı o kadar kez ekrana yazdır

```python
def greet_user():
    user_name=input("Lütfen isminizi giriniz: ") # 1. adım - Kullanıcıdan input fonksiyonu ile isim alalım
    frequency=int(input("İsminizi kaç kez yazdırmak istiyorsunuz?")) # Kullanıcıdan ismini kaç kez konsola yazdırmak istediğine dair bilgi alıyoruz. Burada input string değer alırken int ile o değeri sayıya çeviriyoruz ki for döngüsüne atabilelim.
    print() # çıktımızın güzel gözükmesi için araya 1 satır boşluk bırakıyoruz. 
    for i in range(frequency): # Kullanıcıdan alınan sayı kadar tekrar edecek şekilde print ifadesini konsola yazdırıyoruz. 
        print(f"Merhaba, {user_name}!")

greet_user()
```
