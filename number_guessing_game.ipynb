### Proje 3-Number Guessing Game

Yapılacaklar
* Bilgisayar 1-100 arasında rastgele bir sayı tutar
* Kullanıcı tahmin girer
* "Daha büyük" veya "daha küçük" ipucu verilir
* Bilene kadar devam eder, kaç tahminde bildiğini gösterir

```python
import random # rastgele sayı üretimi için gereken kütüphaneyi import ediyoruz.
def number_guessing():
    random_number = random.randint(1,100)
    guessing_number = 0
    while True:
        try:
            user_prediction = int(input("Sayıyı tahmin edin (1-100): "))
            guessing_number += 1
            if user_prediction < random_number:
                print("Daha büyük bir sayı girin.")
            elif user_prediction > random_number:
                print("Daha küçük bir sayı girin.")
            else:
                print(f"Tebrikler! Sayıyı {guessing_number} tahminde buldunuz.")
                break
        except ValueError:
            print("Hata: 1-100 arası tam sayı giriniz.")
```

```python
number_guessing()
```
