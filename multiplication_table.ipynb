### Proje 4-Multiplication Table

Yapılacaklar
* Kullanıcıdan bir sayı al
* O sayının 1'den 10'a kadar çarpım tablosunu ekrana yazdır

```python
def multiplication_table():
    try:
        user_number = int(input("Bir sayı girin: "))
        print()
        for i in range(1,11):
            print(f"{user_number} x {i:<2} = {user_number * i}")
    except ValueError:
        print("Hata: Tam sayı girişi yapınız.")
```

```python
multiplication_table()
```
