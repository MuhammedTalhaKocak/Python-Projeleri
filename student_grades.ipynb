### Proje 8- Student Grades

```python
def student_grade_system():
    students = {} # öğrencilerin notlarınının kaydedileceği boş sözlük
    while True: # koşul doğru iken
        try: # alttaki kod bloğunu çalıştırmayı dene hata alırsak program çökmeyecek, olası bir hata durumunda except bloğu devreye girecek
            print("Öğrenci eklemek için 1'e basınız.")
            print("Öğrenci silmek için 2'ye basınız.")
            print("Tüm öğrencileri listelemek için 3'e basınız.")
            print("Ortalama hesaplamak için 4'e basınız.")
            print("Çıkmak için 5'e basınız.")
            user_choice = int(input("Lütfen 1-5 arasında bir sayı giriniz: ")) # kullanıcıdan programın dinamikleşmesi için sayı seçimi isteme
            if user_choice == 1: # kullanıcının seçimi 1 ise aşağısı çalışsın.
                student_name = input("Öğrencinin adını giriniz: ")
                student_grade = int(input("Öğrencinin almış olduğu notu giriniz: "))
                if student_name in students: # kullanıcının girdiği isim mevcutsa kullanıcıya aşağıdaki hata mesajı iletilecek
                    print(f"Hata: {student_name} isminde farklı bir öğrenci daha var.")
                else:
                    students[student_name] = student_grade
            elif user_choice == 2: # kullanıcının seçimi 2 ise aşağısı çalışsın
                student_name = input("Listeden silmek istediğiniz öğrencinin adını giriniz: ")
                if student_name not in students: # kullanıcının girdiği öğrenci listede değilse kullanıcıya bilgilendirici hata mesajı verilsin ve 
                    print(f"{student_name} isimli öğrenci listede bulunmamaktadır.")
                else:
                    del students[student_name] # girilen öğrenci mevcutsa sözlükten sil ve kullanıcıyı bilgilendir. 
                    print(f"{student_name} isimli öğrenci listeden silinmiştir.")
            elif user_choice == 3: # kullanıcının seçimi 3 ise aşağısı çalışsın.
                for ogrenci, puan in students.items(): # Çıktının okunabilirliği açısından her bir öğrenci puanıyla beraber alt alta yazılsın.
                    print(f"{ogrenci}: {puan}")
            elif user_choice == 4: # kullanıcının seçimi 4 ise aşağısı çalışsın
                if students:
                    print(f"Öğrencilerin not ortalaması: {sum(students.values()) / len(students)}")
                else: # boş sözlükte ortalama hesaplanması yapılmaya çalışılırsa aşağıdaki kod bloğu ile kullanıcı bilgilendirilsin.
                    print("Hata: Boş sözlük üzerinden ortalama hesaplanamaz.")
            elif user_choice == 5: # kullanıcı programdan çıkmak istediğinde break komutu kullanılarak döngü sona erdirilsin.
                print("Programımız sonlamıştır.")
                break
        except ValueError: # try bloğunda kullanıcıdan veri girişi istenirken kullanıcının girdiği değerin integer'a dönüştürülmesinde sıkıntı yaşanırsa program
            # çökmeden kullanıcıya bilgi verilecek.
            print("Hata: 1-5 arasında bir tam sayı giriniz.")
```

```python
student_grade_system()
```
