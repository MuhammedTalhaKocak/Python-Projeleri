### Proje 6- Word Counter

Yapılacaklar
* Bir txt dosyası oluştur(içine birkaç cümle yaz). Sonra:
* Dosyayı oku
* Küçük harfe çevir
* Kelimelere ayır
* Her kelimenin kaç kez geçtiğini say
* Sonucu ekrana yazdır

```python
try:
    with open("text.txt", "w") as f:
        f.write("hello world hello python world python python\n")
        f.write("İSTANBUL YİRMİ DOKUZ MAYIS ÜNİVERSİTESİ")
    with open("text.txt", "r") as dosya:
        metin = dosya.read().lower()
        kelimeler = metin.split()
    kelime_sayac = {}
    for kelime in kelimeler:
        if kelime in kelime_sayac:
            kelime_sayac[kelime] += 1
        else:
            kelime_sayac[kelime] = 1
    for kelime, sayi in kelime_sayac.items():
        print(f"{kelime}: {sayi}")
except FileNotFoundError:
     print("Hata: Belirtilen dosya bulunamadı.")   
```
