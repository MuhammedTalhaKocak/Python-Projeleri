### Proje 5-FizzBuzz

Yapılacaklar
* Kullanıcıdan bir sayı al
* 1'den o sayıya kadar say
* 3'e bölünebiliyorsa → "Fizz"
* 5'e bölünebiliyorsa → "Buzz"
* İkisine de bölünebiliyorsa → "FizzBuzz"
* Hiçbirine bölünmüyorsa → sayıyı yazdır

```python
def fizz_buzz():
    try:
        user_number = int(input("Bir sayı girin: "))
        if user_number > 0:
            print()
            for i in range(1,user_number+1):
                if i % 15 == 0:
                    print("FizzBuzz")
                elif i % 5 == 0:
                    print("Buzz")
                elif i % 3 == 0:
                    print("Fizz")
                else:
                    print(i)
        else:
            print("Hata: Girdiğiniz sayı bir pozitif sayı olmalıdır.")
    except ValueError:
        print("Hata: Lütfen bir tam sayı giriniz.")    
```

```python
fizz_buzz()
```
