### Proje 7-Shopping List

```python
def shopping():
    shopping_list = [] # kullanıcının arabası (bizim örneğimizde liste) en başta boş
    while True: # kullanıcı çıkmak istemedikçe döngü devam etmeli.
        try: # kodumuzda çıkabilecek olası hatalara karşı programın çökmesini engellemeliyiz.
            print("Alışveriş Listesi")
            print("1. Ürün ekle")
            print("2. Ürün sil")
            print("3. Listeyi görüntüle")
            print("4. Çıkış")
            user_choice = int(input("Lütfen yukarıdaki listeden 1-4 arasında seçiminizi yapınız.")) # kullanıcının seçimi alınıyor. Program dinamik hale geliyor.
            if user_choice == 1:
                urun = input("Lütfen eklemek istediğiniz ürünü yazınız: ")
                shopping_list.append(urun)
            elif user_choice == 2:
                silinecek_urun = input("Lütfen silmek istediğiniz ürünü yazınız: ")
                if silinecek_urun in shopping_list:
                    shopping_list.remove(silinecek_urun)
                    print(f"{silinecek_urun} alışveriş listenizden kaldırıldı.")
                else:
                    print("Hata: Silmek istediğiniz ürün alışveriş listesinde bulunamadı.")
            elif user_choice == 3:
                for urun in shopping_list:
                    print(f"- {urun}")
            elif user_choice == 4:
                break
        except ValueError: # kullanıcının yaptığı bir hatada kullanıcıya bilgilendirici bir metin sunmalıyız. 
            print("Hata: lütfen 1-4 arasında bir tam sayı giriniz.")
            continue
```
